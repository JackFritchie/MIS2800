## Q1
1) Convert data range (starting on Row 11) to a table
2) Filter to the Cincinnati Bengals (CIN) team
3) Filter for the status of "Out"
4) Sort by Player in Ascending Order
5) Use the drop down to list the four players. that are out


```python
# Your file contains repeated header rows inside the data; remove them:
df1 = df[df["Player"] != "Player"].copy()

# Connect to SQLite (in-memory for class demo; you can save to Drive if you want)
con = sqlite3.connect(":memory:")

# Create the SQL table (equivalent to "convert range to a table")
df1.to_sql("injury_report", con, if_exists="replace", index=False)

df1.head()
```

```python
query = """
SELECT Player
FROM injury_report
WHERE Tm = 'CIN'
  AND Status = 'Out'
ORDER BY Player ASC;
"""

result = pd.read_sql_query(query, con)
result
```

## Q2
**Instructions:
1) Convert data range (start on row 11) to a table
2) Filter the Position to WR
3) Filter the Injury Comment to Illness
4) Add a total row to the table (tip: use the search feature at the top by searching for "total row")
5) How many players are out to illness?**

```python
# Remove repeated header rows inside the file
df2 = df[df["Player"] != "Player"].copy()

con = sqlite3.connect(":memory:")
df2.to_sql("injury_report", con, if_exists="replace", index=False)
```

```python
#List the WR players with Illness + Out (sorted)
query_list = """
SELECT Player, Tm, Pos, Status, [Injury Comment]
FROM injury_report
WHERE Pos = 'WR'
  AND [Injury Comment] = 'Illness'
  AND Status = 'Out'
ORDER BY Player ASC;
"""
pd.read_sql_query(query_list, con)
```

```python
# â€œTotal rowâ€ (how many players are out due to illness?)
query_count = """
SELECT COUNT(DISTINCT Player) AS out_due_to_illness_WR
FROM injury_report
WHERE Pos = 'WR'
  AND [Injury Comment] = 'Illness'
  AND Status = 'Out';
"""
pd.read_sql_query(query_count, con)
```

```python
query_list = """
SELECT Player, Tm, Status, [Injury Comment]
FROM injury_report
WHERE [Injury Comment] = 'Concussion'
  AND Status = 'Out'
ORDER BY Player ASC;
"""
pd.read_sql_query(query_list, con)
```
